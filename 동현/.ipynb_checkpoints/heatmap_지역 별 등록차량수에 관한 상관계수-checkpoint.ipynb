```python
pre2 = pd.read_csv("data/additional.csv", encoding = "utf-8")
pre2.head()
pre2.info()
pre2.drop("Unnamed: 0", axis = 1, inplace = True)
pre2.head()
pre3 = pre2.drop(["단지코드","임대건물구분","임대료","노선수","버스","노인청소년합"], axis=1)
pre3.columns
pre3=pre3[['총세대수', '지역', '전용면적', '전용면적별세대수', '빈집', '대중교통', '주차면수', '등록차량수']]
```

```python
pre3 = pre2.drop(["단지코드","임대건물구분","임대료","노선수","버스","노인청소년합"], axis=1)
pre3.columns
pre3=pre3[['총세대수', '지역', '전용면적', '전용면적별세대수', '빈집', '대중교통', '주차면수', '등록차량수']]
```

```python
sido = ['경상남도', '대전광역시', '경기도', '전라북도', '강원도', '광주광역시', '충청남도', '부산광역시',
      '제주특별자치도', '울산광역시', '충청북도', '전라남도', '경상북도', '대구광역시', '서울특별시',
    '세종특별자치시']

area_dict = {}
for i in range(len(sido)):
     area_dict[sido[i]] = pre3[pre3["지역"]== sido[i]]

    
def fu (a) :
    re = area_dict[a].drop("지역",axis = 1)
    r = re.corr()
    cor = r[["등록차량수"]]
    cor.columns = [a + "_등록차량수"]
    return cor
```

```python
fu ('경상남도')
```

```python
result = fu ('경상남도')

for i in sido:
    result = pd.concat([result, fu (i)], axis =1)
    
    
result
```

```python
result2 = result.drop(result.columns[16], axis=1)
result2
```

```python
result2 = result2.iloc[:, [j for j, c in enumerate(result2.columns) if j != 0]]
```

```python
result2 = result2.transpose()
result2
```

```python
plt.figure(figsize=(20,5))
hmap = sns.heatmap(result2, cmap = 'plasma', annot = True)
hmap.set_title("지역별 등록차량수에 관한 상관계수", fontsize=15)
plt.show()
```
